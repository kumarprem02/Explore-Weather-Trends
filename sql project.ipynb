# Gathering:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
city_data=pd.read_csv('city_data.csv')
global_data=pd.read_csv('global_data.csv')
```

```python
city_clean=city_data.copy()
global_clean=global_data.copy()
```

# Comparison of Indian cities temperature with global temperature:

# Global Tempereature line chart across the years:

```python
plt.figure(figsize=(20,10));
plt.plot(global_clean.year,global_clean.avg_temp);
plt.xlabel('Year',fontsize=20);
plt.ylabel('average temperature in °C',fontsize=20);
plt.title('Global temperature in °C across the years',fontsize=20);
```

# Bangalore city Temperature across the years:

```python
plt.figure(figsize=(20,10));
plt.plot(city_clean[city_clean['city']=='Bangalore']['year'],city_clean[city_clean['city']=='Bangalore']['avg_temp']);
plt.xlabel('Year',fontsize=20);
plt.ylabel('Average Temperature in °C',fontsize=20);
plt.title('Bangalore city temperature in °C across the years',fontsize=20);
```

# Bangalore city Temperature v/s Global Temperature:


```python
plt.figure(figsize=(20,10));
plt.plot(city_clean[city_clean['city']=='Bangalore']['year'],city_clean[city_clean['city']=='Bangalore']['avg_temp']);
plt.plot(global_clean.year,global_clean.avg_temp);
plt.xlabel('Year',fontsize=20);
plt.ylabel('Average Temperature in °C',fontsize=20);
plt.title('Bangalore city temperature v/s global temperature in °C',fontsize=20);
```

# Hyderabad city Temperature v/s Global Temperature:
